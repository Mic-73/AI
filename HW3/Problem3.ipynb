```python
## DFS

def DFS(graph, start, goal, visited=None):

    # Create visited set
    if visited is None:
        visited = set()

    # Add start node to visited set
    visited.add(start)

    # Return case where start and goal are equal
    if start == goal:
        return True

    # Loop through neighbors until all are visited
    for neighbor in graph.get(start, []):
        if neighbor not in visited:
            # Recursive call for each new neighbor visited
            if DFS(graph, neighbor, goal, visited):
                return True

    # Return false if goal node was not reached from start node
    return False
```
